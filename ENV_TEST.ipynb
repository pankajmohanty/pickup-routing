```python
import numpy as np
N = 3
M = N*N
INF = 100*7
Data = np.ones((M,M))*INF
for i in range(M):
    print(i)
    Data[i,i]=0
    if(i%N==N-1):  # Right
        if(i != M-1):
            Data[i, i+N] = 1 
        else:
            pass

    elif(i+1>N*(N-1) and i!=M-1): # Down
        Data[i, i+1] = 1 

    else:             # Others
        Data[i, i+1] = 1
        Data[i, i+N] = 1 

    Data[:, i] = np.transpose(Data[i, :]) 

print(Data)
```
